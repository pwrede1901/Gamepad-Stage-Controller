```python
import time
import serial

from homeZ import homeZ
from getpositionZ import get_position_z
from moveZ import moveZ

# Initialize serial instance
com_port = 'COM3'
baud_rate = 115200

try:
    ser = serial.Serial(
        port=com_port,
        baudrate=baud_rate,
        parity=serial.PARITY_NONE,
        stopbits=serial.STOPBITS_ONE,
        bytesize=serial.EIGHTBITS,
        timeout=2,
        xonxoff=False,
        rtscts=False,
        dsrdtr=False
    )

    # Additional settings for line termination (Terminator in MATLAB)
    ser.eol = b'\n'  # Set the line ending to LF (line feed)

    # Check if the port is open
    if ser.is_open:
        print(f"Connected to {com_port} at {baud_rate} baud.")

    # Move Z
    return_val = homeZ(ser)
    pos = 40  # final position (mm)
    vel = 25  # velocity
    acc = 0.3  # acceleration
    return_val = moveZ(ser, pos, vel, acc, 0)
    position = get_position_z(ser)
    print(f"Current position: {position} mm")
    return_val = homeZ(ser)

finally:
    # Close the serial port to release resources
    if ser.is_open:
        ser.close()
        print("Serial port closed.")
```

```python
import time
import serial

from homeZ import homeZ
from getpositionZ import get_position_z
from moveZ import moveZ

# Initialize serial instance
com_port = 'COM3'
baud_rate = 115200

ser = None  # Initialize the serial port variable

try:
    # Check if the port is already open
    if not ser or not ser.is_open:
        ser = serial.Serial(
            port=com_port,
            baudrate=baud_rate,
            parity=serial.PARITY_NONE,
            stopbits=serial.STOPBITS_ONE,
            bytesize=serial.EIGHTBITS,
            timeout=2,
            xonxoff=False,
            rtscts=False,
            dsrdtr=False
        )

        # Additional settings for line termination (Terminator in MATLAB)
        ser.eol = b'\n'  # Set the line ending to LF (line feed)

        print(f"Connected to {com_port} at {baud_rate} baud.")

    # Move Z
    return_val = homeZ(ser)
    pos = 40  # final position (mm)
    vel = 25  # velocity
    acc = 0.3  # acceleration
    return_val = moveZ(ser, pos, vel, acc, 0)
    position = get_position_z(ser)
    print(f"Current position: {position} mm")
    return_val = homeZ(ser)

finally:
    # Close the serial port to release resources
    if ser and ser.is_open:
        ser.close()
        print("Serial port closed.")
```
